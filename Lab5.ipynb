```python
import pandas as pd
import numpy as np
import seaborn as sns

die3 = []
#Roll a 3 sided die with the following face values 1000 times and calculate the mean and standard deviation of the rolls
for i in range(10000):
    i = np.random.choice([5,6,7], p=[1/3, 1/3, 1/3])
    die3.append(i)
print(die3)

sum_die = sum(die3)
mean = sum_die/10000
print(mean)

print(np.std(die3))
```

```python
die5 = []

#Roll a 5 sided die with the following face values 1000 times and calculate the mean and standard deviation
for i in range(10000):
    i = np.random.choice([3,5,7,9,11], p=[1/5, 1/5, 1/5, 1/5, 1/5])
    die5.append(i)

sum_die = sum(die5)
mean = sum_die/10000
print(mean)

print(np.std(die5))
```

```python
rolls = np.concatenate([die3, die5])
die = np.concatenate([3*np.ones(10000), 5*np.ones(10000)])
dice = pd.DataFrame({'rolls': rolls, 'die': die})

dice
```

```python
#Create histogram for 3 and 5 sided die rolls
fig1 = sns.histplot(data = dice, x = 'rolls', hue = 'die', multiple = 'stack')
fig1.set(xlabel = 'Die Values', ylabel = 'Number of Rolls', title = 'Dice Histogram')
```

```python
#Create dataframe containing covid data for Russia and USA
covid = pd.read_csv("covid.csv")
covid['date'] = pd.to_datetime(covid['date'], format = '%Y-%m-%d')
inds = (covid['location'] == "Russia") | (covid['location'] == "United States")
covid_us_ru = covid[inds]

```

```python
#lineplot of number of new cases reported per day in Russia and USA 
fig2 = sns.lineplot(data = covid_us_ru, x = 'date', y = 'new_cases', hue = 'location')
fig2.set(xlabel = 'Date', ylabel = 'New Cases', title = 'New Cases in USA and Russia')
```
