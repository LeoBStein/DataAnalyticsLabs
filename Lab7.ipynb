```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

workers = pd.read_csv("workers.csv")
online_ratings = pd.read_csv("online_ratings.csv")

online_ratings_melted=pd.melt(online_ratings,id_vars=['id'], var_name='year', value_name='rating')
online_ratings_melted
```

```python
workers_melted=pd.melt(workers, id_vars=['id'], var_name='availability', value_name='recommendation')

workers_melted
```

```python
workers_melted = workers_melted.dropna()
workers_melted
```

```python
workers_all = pd.merge(left=workers_melted, right=online_ratings_melted, on='id', how='outer')
workers_all
```

```python
workers_now = workers_all[workers_all['availability']=='Now']
workers_now = workers_now.sort_values(by='recommendation', ascending=False)
workers_now
```
