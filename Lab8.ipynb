```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

wayfair = pd.read_csv('wayfair.csv')
amazon = pd.read_csv('amazon.csv')

#only want data from the 200 days in amazon.csv
wayfair200 = wayfair[wayfair['date'].isin(amazon['date'])]
wayfair200 = wayfair200.reset_index()

wayfair200
```

```python
#figure out how much and what data is missing
missing_wayfair200 = wayfair200.isnull().sum()
missing_wayfair200
```

```python
#how many days a promotion is being run
(wayfair200['promotion']== True).sum()
```

```python
#replace missing values in promotion column with most frequent promotion value
wayfair200['promotion'] = wayfair200['promotion'].fillna(wayfair200['promotion'].mode()[0])
wayfair200
```

```python
#for each day missing sales data, replace sales data with data from most recent preceding day that we have sales data for
for i in range((wayfair200['sales'].size)):
    if wayfair200['sales'].isnull()[i]:
        wayfair200['sales'][i] = wayfair200['sales'][i-1]

wayfair200
```

```python
#compute avg of variable sales on day promotion was and was not run
n=0
m=0
sales_sum_promo = 0
sales_sum_nopromo = 0
for i in range((wayfair200['sales'].size)):
    if wayfair200['promotion'][i] == True:
        sales_sum_promo = sales_sum_promo + wayfair200['sales'][i]
        n=n+1
    else:
        sales_sum_nopromo = sales_sum_nopromo + wayfair200['sales'][i]
        m=m+1

sales_mean_promo = sales_sum_promo/n
sales_mean_nopromo = sales_sum_nopromo/m

print(sales_mean_promo)
print(sales_mean_nopromo)

        
```
